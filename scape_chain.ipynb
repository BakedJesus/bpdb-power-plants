```python
pcsv = pd.read_csv('hd_plants_left.csv', index_col=0)
```

```python
pcsv = pcsv[['plant_name', "capacity_mw","plant_name_alt",'fuel_type_x', 'own_x']].fillna('').astype('string')
```

```python
pcsv['search_name'] = pcsv.agg(' '.join, axis=1)
```

```python
instructions = """You are a world class energy researcher, who can do detailed research on any power plant and produce facts based results; 
                    you do not make things up, you will try as hard as possible to gather facts & data to back up the research

                    Given the name of a power plant, your goal is to find and return the year in which the plant established. While doing so, abide by the following rules;
                    1/ You should do enough research to gather as much information as possible about the objective
                    2/ If there are url of relevant links & articles, you will scrape it to gather more information
                    3/ After scraping & search, you should think "is there any new things i should search & scraping based on the data I collected to increase research quality?" If answer is yes, continue; But don't do this more than 3 iteratins
                    4/ You should not make things up, you should only write facts & data that you have gathered
                    5/ In the final output, only return the numerical value of the year of establishment, if not found return None. Do not return any text apart from the year itself"""
#
#
```

```python
pcsv['search_year'] = ""
for i, row in pcsv.iloc[62:].iterrows():
    print("Searching", row["search_name"])
    #pcsv.loc[i, "search_year"] = web_reader.researchAgent(f'which year was {row["search_name"]} power plant established', instructions)
    pcsv.loc[i, "search_year"] = web_reader.researchAgent(f'{row["search_name"]}', instructions, verbose=False)
    pcsv.to_csv('hd_left_gpt_out.csv')
```

```python
import re
def extract_year(text):
    year_pattern = re.compile(r'\b(1[89]\d\d|2[0-9]{3})\b')
    match = year_pattern.search(text)
    return match.group() if match else None

pcsv['extracted_year'] = pcsv['search_year'].apply(extract_year)
pcsv.to_csv('hd_left_gpt.csv')
```
